```python
def get_bigrams_ct(words):
    ch_set = set()
    for word in words:
        ch_set.update(set(word))
    N = torch.zeros()
    b = {}
    for word in words[:1]:
        words_chars = ['.'] + list(word) + ['.']
        for ch1, ch2 in zip(words_chars, words_chars[1:]):
            bigram = (ch1, ch2)
            b[bigram] = b.get(bigram, 0) + 1
```
